```python
import pandas as pd
import sklearn as skl 
import numpy as np
import matplotlib.pyplot as plt 
%store -r cat_encode
%store -r liste_donnee
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer, StandardScaler, MinMaxScaler
```

```python
cat_pipeline = Pipeline([
    ("get_dumies_cat",OneHotEncoder(handle_unknown='ignore'))
    ]
)

num_pipeline1= Pipeline([
    ("tranformation_log",FunctionTransformer(np.log1p, inverse_func=np.exp)),
    ("standartization",StandardScaler())
    ]
)

num_colone = ['year', 'duration_fmt', 'danceability', 'energy','tempo']
cat_column = ['track_name', 'artists', 'region', 'country_code', 'track_genre']
```

```python
processing=ColumnTransformer([
    ("cat_transformation", cat_pipeline, cat_column),
    ("num_transformation" ,num_pipeline1, num_colone)
])
```

```python
data=liste_donnee[0][0]
data = data.replace([np.inf, -np.inf], np.nan)
```
